```python
import json
import numpy as np
import pandas as pd
import plotly.io as pio
import plotly.express as px
```

```python
tun_states=json.load(open("G:\\tun\delegations_raw.geojson", 'r',encoding='utf8'))
tun_states["features"][0]
```

```python
state_id_map = {}
for feature in tun_states["features"]:
    feature["id"] = feature["properties"]["del_id"]
    state_id_map[feature["properties"]["del_fr"]] = feature["id"]
```

```python
df = pd.read_csv("G:\\tun\data_carto_census2014.csv")
df.head()
```

```python
df["firmsScale"] = np.log10(df["firms"])
```

```python
fig = px.choropleth(
    df,
    locations="id",
    geojson=tun_states,
    color="firms",
    hover_data=["firms"],
    color_continuous_scale=px.colors.sequential.Redor,
    title="tunisia firms 2020",
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
